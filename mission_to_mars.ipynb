```python
from bs4 import BeautifulSoup as bs
import pandas as pd
from splinter import Browser
import requests
import pymongo
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from flask import Flask, render_template, redirect
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://redplanetscience.com/'
browser.visit(url)
html = browser.html
soup = bs(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
#NASA Mars News
titles = soup.find('div', class_="content_title")
paragraphs = soup.find('div', class_="article_teaser_body")
news_title = titles.text.strip()
news_p = paragraphs.text.strip()
print(f'{news_title}\n{news_p}')
```

```python
hemisphere_url = "https://marshemispheres.com/"
html = requests.get(hemisphere_url)
soup4 = bs(html.text,"html.parser")
```

```python
hemisphere_img = soup4.find_all('div', {"class":["description",'itemLink product-item']})
```

```python
hemisphere_image_urls = []
img_hold = []
hemisphere_img = soup4.find_all('div', {"class":["description",'itemLink product-item']})
for j in hemisphere_img:
    k = j.find('a',href=True)
    img_hold.append(k['href'])
hemisphere_title = soup4.find_all('h3')
for i in range(len(hemisphere_img)):
    hold = {}
    hold['title'] = hemisphere_title[i].text.strip()
    urlh = hemisphere_url+img_hold[i]
    html = requests.get(urlh)
    souph = bs(html.text,"html.parser")
    img_dl= (souph.find('img', class_='wide-image')).get('src')
    hold['img_url']= hemisphere_url+img_dl
    hemisphere_image_urls.append(hold)
    
hemisphere_image_urls
```
